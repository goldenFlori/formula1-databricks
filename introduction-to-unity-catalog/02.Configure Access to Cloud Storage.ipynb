###Configure Access to Cloud Storage via Unity Catalog

#####Access Cloud Storage

```sql
%fs ls 'abfss://demo@databrickscourseextdl999.dfs.core.windows.net'
```

#####Create External Location

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricks_course_ext_dl999_demo URL 'abfss://demo@databrickscourseextdl999.dfs.core.windows.net' WITH (STORAGE CREDENTIAL `databricks-course-sc`) COMMENT 'External location for the demo container';
```
