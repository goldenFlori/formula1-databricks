### Understanding Delta Lake Transaction Log
Understand the cloud storage directory structure for a delta table

#####1. Create Catalog and Schema for the Demo

**Catalog:** demo  
**Schema:** delta_lake

```sql
CREATE CATALOG IF NOT EXISTS demo
MANAGED LOCATION 'abfss://demo@databrickscourseextdl999.dfs.core.windows.net/'
```

```sql
CREATE SCHEMA IF NOT EXISTS demo.delta_lake
MANAGED LOCATION 'abfss://demo@databrickscourseextdl999.dfs.core.windows.net/delta_lake'
```

#####1. Create a Delta Lake Table

```sql
 CREATE TABLE IF NOT EXISTS demo.delta_lake.companies
 (
    company_name STRING,
    founded_date DATE,
    country STRING
 )
USING DELTA;
```

```sql
DESC EXTENDED demo.delta_lake.companies;
```

#####2. Insert Some Data

```sql
INSERT INTO demo.delta_lake.companies
VALUES ('Apple', '1976-04-01', 'USA')
```

```sql
SELECT * FROM demo.delta_lake.companies;
```
