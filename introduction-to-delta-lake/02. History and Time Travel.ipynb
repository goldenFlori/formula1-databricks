### History and Time Travel

1. Query Delta Lake table history
2. Query previous versions of the data
3. Query data from a specific time
4. Restore data to a specific version

#####1. Query Delta Table History

```sql
DESCRIBE HISTORY demo.delta_lake.companies
```

#####2. Query Data from a Specific Version

```sql
SELECT * FROM demo.delta_lake.companies VERSION AS OF 1
```

#####3. Query Data from a Specific Time

```sql
SELECT * FROM demo.delta_lake.companies TIMESTAMP AS OF '2026-07-11T21:40:42.000+00:00'
```

#####4. Restore Data in the Table to a Specific Version

```sql
RESTORE TABLE demo.delta_lake.companies TO VERSION AS OF 1
```
